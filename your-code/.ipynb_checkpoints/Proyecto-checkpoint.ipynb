```python
casas_f = casas[["OverallQual", "YearBuilt", "YearRemodAdd", "TotalBsmtSF", "GrLivArea", "FullBath", 
            "GarageCars", "GarageArea"]]
```

```python
casas_f.describe().transpose()
```

```python
#Procedemos a estandarizar las variables

#Tenemos que modificar las variables de "YearBuilt" y "YearRemodAdd" para que demuestren la diferencia con 2020.
casas_f["YearBuilt_cod"] = (casas["YearBuilt"]-2020)*(-1)
casas_f["YearRemodAdd_cod"] = (casas["YearRemodAdd"]-2020)*(-1)
```

```python
#Fullbath tiene valores de 0 a 3, vamos a convertir la variable a una dummy que tenga el valor 1 si tiene más de un baño
casas_f["FullBath_cod"] = 0
casas_f.loc[casas_f.FullBath == 2, "FullBath_cod"] = 1
casas_f.loc[casas_f.FullBath == 3, "FullBath_cod"] = 1
```
