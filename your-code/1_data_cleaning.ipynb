```python
# Importar las librerias necesarias

import pandas as pd
import numpy as np
import re
```

```python
# Importar el dataset con la informaciÃ³n completa

data = pd.read_csv('time_series_60min_singleindex.csv')
data.head()
```

```python
# Las siguientes listas tienen los codigos de los paÃ­ses de las columnas del dataset

eu_codes = ['AT', 'BE', 'BG', 'CH', 'CY', 'CZ', 'DE', 'DK', 'EE', 'ES', 'FI', 'FR', 'GB', 'GR', 'HR', 'HU', 'IE', 'IT', 'LT', 'LU', 'LV', 'ME', 'NL', 'NO', 'PL', 'PT', 'RO', 'RS', 'SE', 'SI', 'SK', 'UA']

eu_names = ['Austria', 'Belgium', 'Bulgaria', 'Switzerland', 'Cyprus', 'Czechia', 'Germany', 'Denmark', 'Estonia', 'Spain',
'Finland', 'France', 'GreatBritain', 'Greece', 'Croatia', 'Hungary', 'Ireland','Italy', 'Lithuania', 'Luxembourg', 'Latvia', 'Montenegro', 
'Netherlands', 'Norway', 'Poland', 'Portugal', 'Romania', 'Serbia', 'Sweden', 'Slovenia', 'Slovakia', 'Ukraine']
```

```python
# Ahora se va a generar un filtro de columnas por paÃ­s para separar el dataset por: 1-PaÃ­s y 2-Tipo de generaciÃ³n de energÃ­a

# Diccionario con cÃ³digo y paÃ­s
codes_names_dict = dict(zip(eu_codes, eu_names))

# Lista de columnas de dataset original
columns = data.columns

# Ciclo for para filtrar columnas por cÃ³digo
countries_dict = {}

for code in eu_codes:
    str = code + '.+'
    r = re.compile(str)
    countries_dict[code] = list(filter(r.match, columns))

# Ciclo for para cambiar las llaves de countries_dict de codigo a nombre de pais
for code, name in zip(eu_codes, eu_names):
    countries_dict[f'{name}'] = countries_dict.pop(code)

# Filtrar el dataframe por paÃ­s y guardar en csv
for name in eu_names:
    data.loc[:,countries_dict[name]].to_csv(f'./data_by_country/{name}.csv')
```

```python
# Lista de tipos de generaciÃ³n de energÃ­a y precio
types = ['load','solar','wind','price']
    # load = generaciÃ³n convencional por combustibles y nucleares
    # solar = generaciÃ³n solar
    # wind = generacion por viento


# Ciclo for para separar columnas por tipo
types_dict = {}

for typ  in types:
    str = '.+' + typ + '.+'
    r = re.compile(str)
    types_dict[typ] = list(filter(r.match, columns))
```

```python
# SeparaciÃ³n de columnas por tipo de generaciÃ³n

load = types_dict['load']
r = re.compile(f'.+load_actual.+')
load = list(filter(r.match, load))

solar = types_dict['solar']
r = re.compile(f'.+generation_actual')
solar = list(filter(r.match, solar))

wind = types_dict['wind']
r = re.compile(f'.+generation_actual')
wind = list(filter(r.match, wind))

price = types_dict['price']

print(load[:5])
print(solar[:5])
print(wind[:5])
print(price[:5])
```

```python
# Filtrar el dataframe por paÃ­s y guardar en csv

data.loc[:,load].to_csv('./data_by_type/load_df.csv')
data.loc[:,solar].to_csv('./data_by_type/solar_df.csv')
data.loc[:,wind].to_csv('./data_by_type/wind_df.csv')
data.loc[:,price].to_csv('./data_by_type/price_df.csv')
```
