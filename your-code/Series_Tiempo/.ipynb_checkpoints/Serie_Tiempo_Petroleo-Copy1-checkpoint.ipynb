```python
def viewing_stationarity(df, window=5):
  #Esta funcin genera un plot para una serie de tiempo con su media mvil y su desviacin estndar.


    rolmean = df.rolling(window=window).mean()
    rolstd = df.rolling(window=window).std()

    fig = plt.figure(figsize=(10, 5))
    time_series = plt.plot(df, color='blue', label='original')
    mean = plt.plot(rolmean, color='red', label='media mvil')
    std = plt.plot(rolstd, color='green', label='desviacin estndar mvil')
    plt.legend(loc='best')
    plt.show();
```

```python
class Stationary:
    def __init__(self, significance=0.05):
        self.significance_level = significance
        self.pvalue = None
        self.is_stationary = None

    def ADF(self, values):
        result = adfuller(values, autolag='AIC')

        format_results = pd.Series(result[:4], 
                                    index=['Statistic', 'P-value',
                                          'Num of Lags', 'Observations'])
        for k,v in result[4].items():
              format_results[f'Critical Value {k}'] = v

        return format_results
```

```python
viewing_stationarity(petroleo, window=60)
adf_test = Stationary()
adf_test.ADF(petroleo.values).round(5)
```

```python
petroleo['Closing_value_shift'] = petroleo['Closing Value'].shift(1) # bits = n, bits_shift1 = n-1
petroleo.head()
```

```python
petroleo['Closing_value_diff'] = petroleo['Closing Value'] - petroleo['Closing_value_shift']
petroleo.head()
```

```python
viewing_stationarity(petroleo['Closing_value_diff'].dropna(), window=200)
adf_test = Stationary()
adf_test.ADF(petroleo['Closing_value_diff'].dropna().values).round(6)
```

# Metodo ARIMA/SARIMAX

```python
model=sm.tsa.arima.ARIMA(petroleo['Closing Value'],order=(1,1,1))
history=model.fit()
history.summary()
```

```python
petroleo.tail()
```

```python
petroleo['prediccion']=history.predict(start="2016-02-02",end="2018-09-07",dynamic=True)
petroleo[['Closing Value','prediccion']].plot(figsize=(12,8));
```

```python
model=sm.tsa.arima.ARIMA(petroleo['Closing_value_diff'].dropna(),order=(1,1,1))
model_fit=model.fit()
```

```python
petroleo['prediccion']=model_fit.predict(start="2016-02-02",end="2018-09-07",dynamic=True)
petroleo[['Closing_value_diff','prediccion']].plot(figsize=(12,8))
```

```python
model=sm.tsa.statespace.SARIMAX(petroleo['Closing Value'],order=(1, 1, 1),seasonal_order=(1,1,1,12))
results=model.fit()
```

```python
petroleo['prediccion']=results.predict(start="2016-02-02",end="2018-09-07",dynamic=True)
petroleo[['Closing Value','prediccion']].plot(figsize=(12,8))
```
