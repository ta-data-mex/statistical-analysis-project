```python
def viewing_stationarity(df, window=5):
  #Esta funcin genera un plot para una serie de tiempo con su media mvil y su desviacin estndar.


    rolmean = df.rolling(window=window).mean()
    rolstd = df.rolling(window=window).std()

    fig = plt.figure(figsize=(10, 5))
    time_series = plt.plot(df, color='blue', label='original')
    mean = plt.plot(rolmean, color='red', label='media mvil')
    std = plt.plot(rolstd, color='green', label='desviacin estndar mvil')
    plt.legend(loc='best')
    plt.show();
```

```python
class Stationary:
    def __init__(self, significance=0.05):
        self.significance_level = significance
        self.pvalue = None
        self.is_stationary = None

    def ADF(self, values):
        result = adfuller(values, autolag='AIC')

        format_results = pd.Series(result[:4], 
                                    index=['Statistic', 'P-value',
                                          'Num of Lags', 'Observations'])
        for k,v in result[4].items():
              format_results[f'Critical Value {k}'] = v

        return format_results
```

```python
viewing_stationarity(petroleo_anual, window=60)
adf_test = Stationary()
adf_test.ADF(petroleo_anual.values).round(5)
```

```python
petroleo_anual['Closing_value_shift'] = petroleo_anual['Closing Value'].shift(1) # bits = n, bits_shift1 = n-1
petroleo_anual.head()
```

```python
petroleo_anual['Closing_value_diff'] = petroleo_anual['Closing Value'] - petroleo_anual['Closing_value_shift']
petroleo_anual.head()
```

```python
viewing_stationarity(petroleo_anual['Closing_value_diff'].dropna(), window=200)
adf_test = Stationary()
adf_test.ADF(petroleo_anual['Closing_value_diff'].dropna().values).round(6)
```

# Metodo ARIMA/SARIMAX

```python
model=sm.tsa.arima.ARIMA(petroleo_anual['Closing Value'],order=(1,1,1))
history=model.fit()
history.summary()
```

```python
petroleo_anual.tail()
```

```python
petroleo_anual['prediccion']=history.predict(start="2014-12-31",end="2018-12-31",dynamic=True)
petroleo_anual[['Closing Value','prediccion']].plot(figsize=(12,8));
```

```python
model=sm.tsa.arima.ARIMA(petroleo_anual['Closing_value_diff'].dropna(),order=(1,1,1))
model_fit=model.fit()
model_fit.summary()
```

```python
petroleo_anual['prediccion']=model_fit.predict(start="2014-12-31",end="2018-12-31",dynamic=True)
petroleo_anual[['Closing_value_diff','prediccion']].plot(figsize=(12,8));
```

```python
model=sm.tsa.statespace.SARIMAX(petroleo_anual['Closing Value'],order=(1, 1, 1),seasonal_order=(1,1,1,15))
results=model.fit()
results.summary()
```

```python
petroleo_anual['prediccion']=results.predict(start="2014-12-31",end="2018-12-31",dynamic=True)
petroleo_anual[['Closing Value','prediccion']].plot(figsize=(12,8))
```

```python
results.plot_diagnostics()
plt.show()
```

```python
rcParams['figure.figsize'] = 30, 15

pred_uc = results.get_forecast(steps=2)
pred_ci = pred_uc.conf_int()
ax = petroleo_anual.plot(label='Observado', figsize=(30, 15))
pred_uc.predicted_mean.plot(ax=ax, label='Prediccin futura')
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.25)
ax.set_xlabel('Fecha')
ax.set_ylabel('Closing value')
plt.legend()
plt.show()
```
