```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
casas.dtypes
```

```python
x_casas = sm.add_constant(casas[["LotArea", "OverallCond", "OverallQual","TotalBsmtSF", "YearBuilt","YearRemodAdd","GrLivArea", "GarageCars"]])
y_casas = casas["SalePrice"]
```

```python
casas.LotArea.corr(casas.SalePrice, method="pearson")
```

```python
casas.OverallCond.corr(casas.SalePrice, method="pearson")
```

```python
housemodel = sm.OLS(y_casas, x_casas).fit()
predictions = housemodel.predict()

print_model = housemodel.summary()
print_model
```
